```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Muat dataset
file_path = 'data_cuaca.csv'  # Sesuaikan lokasi file Anda
data = pd.read_csv(file_path, sep=';')

# Pastikan kolom numerik
data['ch'] = pd.to_numeric(data['ch'], errors='coerce')
data['temp_rata-rata'] = pd.to_numeric(data['temp_rata-rata'], errors='coerce')

# Filter data untuk curah hujan yang masuk akal (misal < 5000 mm)
filtered_data = data[data['ch'] < 5000]

# Scatter plot dengan log scale pada sumbu X
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_data, x='ch', y='temp_rata-rata', alpha=0.7, color='orange')

# Set log scale pada sumbu X
plt.xscale('log')

# Tambahkan judul dan label
plt.title('Scatter Plot: Curah Hujan (ch) vs Suhu Rata-rata (Log Scale)', fontsize=14)
plt.xlabel('Curah Hujan (mm, Log Scale)', fontsize=12)
plt.ylabel('Suhu Rata-rata (\u00b0C)', fontsize=12)
plt.grid(alpha=0.3, which='both')

# Tampilkan plot
plt.show()
```

```python
# Import library yang diperlukan
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset
file_path = 'data_cuaca.csv'  # Ganti dengan lokasi file dataset kamu
data = pd.read_csv(file_path, sep=';')

# Memilih kolom numerik yang relevan untuk analisis korelasi
numerical_data = data[["temp_min", "temp_max", "temp_rata-rata", "lembab_rata-rata", "ch", "cahaya_jam"]]

# Menghitung matriks korelasi
correlation_matrix = numerical_data.corr()

# Membuat heatmap menggunakan Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Pairwise Correlation")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Muat dataset (ganti 'data_cuaca.csv' dengan lokasi file Anda)
data = pd.read_csv('data_cuaca.csv', sep=';')

# Pastikan kolom numerik sesuai
data['cahaya_jam'] = pd.to_numeric(data['cahaya_jam'], errors='coerce')
data['temp_max'] = pd.to_numeric(data['temp_max'], errors='coerce')

# Membuat scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='cahaya_jam', y='temp_max', alpha=0.7, color='green')

# Tambahkan judul dan label
plt.title('Scatter Plot: Cahaya Jam vs Suhu Maksimum', fontsize=14)
plt.xlabel('Cahaya Jam (Jam)', fontsize=12)
plt.ylabel('Suhu Maksimum (°C)', fontsize=12)
plt.grid(alpha=0.3)

# Tampilkan plot
plt.show()
```
